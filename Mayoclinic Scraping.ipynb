```python
table={'Name':[],'Overview':[],'Symptoms':[],'When to see a doctor':[],'Causes':[],'Risk factors':[],'Complications':[]
      ,'Prevention':[]}
```

```python
for i in urls:
    print(i)  
    try:
        url = i
        response = requests.get(url)
        html = response.text
        soup = BeautifulSoup(html,'html.parser')
        table['Name'].append(soup.find_all('h1')[0].text.strip())
        if len(soup.find_all('h2',string=re.compile('Overview')))==0:
            table['Overview'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Overview'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Overview'].append("".join(plot))
        if len(soup.find_all('h2',string=re.compile('Symptoms')))==0:
            table['Symptoms'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Symptoms'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Symptoms'].append("".join(plot))
        if len(soup.find_all('h3',string=re.compile('When to see a doctor')))==0:
            table['When to see a doctor'].append('')
        else:
            mark=soup.find('h3',string=re.compile('When to see a doctor'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['When to see a doctor'].append("".join(plot))
        if len(soup.find_all('h2',string=re.compile('Causes')))==0:
            table['Causes'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Causes'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h2',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Causes'].append("".join(plot))
        if len(soup.find_all('h2',string=re.compile('Risk factors')))==0:
            table['Risk factors'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Risk factors'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Risk factors'].append("".join(plot))
        if len(soup.find_all('h2',string=re.compile('Complications')))==0:
            table['Complications'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Complications'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Complications'].append("".join(plot))
        if len(soup.find_all('h2',string=re.compile('Prevention')))==0:
            table['Prevention'].append('')
        else:
            mark=soup.find('h2',string=re.compile('Prevention'))
            plot=[]
            for elt in mark.find_next_siblings():
                if re.search('h[0-9]',elt.name):
                    break
                if hasattr(elt, "text"):
                    plot.append(elt.text.strip())
            table['Prevention'].append("".join(plot))
    except IndexError:
        err_url.append(i)
```
