```python
disease5=pd.read_excel('Diseases_5.xlsx')
```

```python
disease5
```

```python
Disease_new=[]
for i in disease5.Disease:
    a=i
    if re.search('#',a):
        a=re.sub('#','',a)
    if re.search('\*\*',a):
        a=re.sub('\*\*','',a)
    if re.search('!',a):
        a=re.sub('!','',a)
    Disease_new.append(a.strip())
Disease_new
```

```python
disease5['Disease_new']=Disease_new
disease5.drop(['Disease'],axis=1,inplace=True)
disease5
```

```python
d1_d2_d3_d4_d5_outer=pd.merge(d1_d2_d3_d4,disease5,how='outer',left_on='New_Disease_col3',right_on='Disease_new')
d1_d2_d3_d4_d5_outer
```

```python
d1_d2_d3_d4_d5_inner=pd.merge(d1_d2_d3_d4,disease5,how='inner',left_on='New_Disease_col3',right_on='Disease_new')
d1_d2_d3_d4_d5_inner=d1_d2_d3_d4_d5_inner.drop_duplicates().reset_index()
d1_d2_d3_d4_d5_inner.drop(['index'],axis=1,inplace=True)
```

```python
d1_d2_d3_d4_d5=d1_d2_d3_d4_d5_inner.append(d1_d2_d3_d4_d5_outer)
d1_d2_d3_d4_d5=d1_d2_d3_d4_d5.fillna('')
```

```python
new_disease_final=[]
for i in range(0,len(d1_d2_d3_d4_d5['New_Disease_col3'])):
    if d1_d2_d3_d4_d5.iloc[i,16]=='':
        new_disease_final.append(d1_d2_d3_d4_d5.iloc[i,21])
    else:
        new_disease_final.append(d1_d2_d3_d4_d5.iloc[i,16])
new_disease_final
```

```python
d1_d2_d3_d4_d5['Disease name Final']=new_disease_final
```

```python
d1_d2_d3_d4_d5.drop(['New_Disease_col3','Disease_new'],axis=1,inplace=True)
```

```python
d1_d2_d3_d4_d5.to_csv('d1_2_3_4_5.csv',index=False)
```

```python
df_kgl=pd.read_csv('dataset.csv')
```

```python
df_kgl=df_kgl.drop_duplicates()
df_kgl
```

```python
df_kgl=df_kgl.fillna('')
```

```python
kgl={}
for i in df_kgl['Disease'].unique():
    kgl[i]=[]
kgl
```

```python
for i in df_kgl['Disease'].unique():
    for j in range(0,len(df_kgl)):
        if df_kgl.iloc[j,0]==i:
            for k in range(1,18):
                kgl[i].append(df_kgl.iloc[j,k].strip())
```

```python
symptoms=[]
for i in kgl.keys():
    a=list(set(kgl[i]))
    b=",".join(a)
    symptoms.append(b)
symptoms
```

```python
df_kgl_final=pd.DataFrame()
```

```python
df_kgl_final['Disease']=kgl.keys()
df_kgl_final['Symptoms']=symptoms
df_kgl_final
```

```python
df_desc=pd.read_csv('symptom_Description.csv')
```

```python
df_desc=df_desc.drop_duplicates()
df_desc
```

```python
df_kgl_desc=pd.merge(left=df_kgl_final,right=df_desc,on='Disease',how='outer')
df_kgl_desc
```

```python
df_prec=pd.read_csv('symptom_precaution.csv')
df_prec
```

```python
df_kgl_desc_prec=pd.merge(left=df_kgl_desc,right=df_prec,on='Disease',how='outer')
df_kgl_desc_prec=df_kgl_desc_prec.drop_duplicates()
df_kgl_desc_prec
```

```python
df_seve=pd.read_csv('Symptom-severity.csv')
df_seve.rename({'Symptom':'Symptoms'},axis=1,inplace=True)
df_seve
```

```python
dd={'Disease':[],'total_weight':[]}
for i in range(0,len(df_kgl_desc_prec)):
    total=0
    for k in range(0,len(df_seve)):
        if df_seve.iloc[k,0] in str(df_kgl_desc_prec.iloc[i,1]).split(','):
            total=total+df_seve.iloc[k,1]
    dd['Disease'].append(df_kgl_desc_prec.iloc[i,0])
    dd['total_weight'].append(total)
dd
```

```python
df_seve_final=pd.DataFrame(dd)
df_seve_final
```

```python
df_kgl_desc_prec_seve=pd.merge(left=df_kgl_desc_prec,right=df_seve_final,on='Disease',how='inner')
df_kgl_desc_prec_seve
```
