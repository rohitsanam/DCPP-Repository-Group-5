```python
import pandas as pd
import numpy as np
import re
```

# Cleaning Mayo clinic

```python
df_mayo=pd.read_csv('mayoclinic_data_cleaned.csv')
```

```python
#removing serivce error rows
df_mayo=df_mayo[df_mayo['Name'] != 'Server Error']
df_mayo=df_mayo.reset_index()
```

```python
df_mayo.head()
```

```python
df_mayo.drop(['Overview','index'],axis=1,inplace=True)
```

```python
#cleaning 
Symptoms_new=[]
for i in df_mayo['Symptoms']:
    if type(i)==str and re.search('Open pop-up dialog box',i):
        res = []
        for j in i.split('\n'):
            if j not in res:
                res.append(j.strip())
        rew_new=[]
        for i in res:
            if i == 'Open pop-up dialog box' or i=='' or i=='Close':
                continue
            else:
                rew_new.append(i)
        Symptoms_new.append("\n".join(rew_new))
    elif type(i)==str:
        Symptoms_new.append(i)
    else:
        Symptoms_new.append('')
Symptoms_new
```

```python
df_mayo['Symptoms_new']=Symptoms_new
```

```python
#cleaning 
When_to_see_a_doctor_new=[]
for i in df_mayo['When to see a doctor']:
    if type(i)==str and re.search('Open pop-up dialog box',i):
        res = []
        for j in i.split('\n'):
            if j not in res:
                res.append(j.strip())
        rew_new=[]
        for l in res:
            if l == 'Open pop-up dialog box' or l=='' or l=='Close' or re.search('[eE]mail',l):
                continue
            else:
                rew_new.append(l)
        When_to_see_a_doctor_new.append("\n".join(rew_new))
    elif type(i)==str:
        rew_new_new=[]
        for k in i.split('\n'):
            if re.search('[eE]mail',k):
                continue
            else:
                rew_new_new.append(k)
        When_to_see_a_doctor_new.append("\n".join(rew_new_new))
    else:
        When_to_see_a_doctor_new.append('')
When_to_see_a_doctor_new
```

```python
df_mayo['When_to_see_a_doctor_new']=When_to_see_a_doctor_new
```

# Loading and cleaning Kegg

```python
df_kegg=pd.read_csv('kegg disease.csv')
```

```python
df_kegg.head()
```

```python
name_caps=[]
for i in df_kegg['Name']:
    name_caps.append(i.upper())
df_kegg['Name_caps']=name_caps
df_kegg=df_kegg[['Entry', 'Name_caps', 'Description', 'Category', 'Brite', 'Carcinogen',
       'Drug', 'Gene']]
```

```python
df_kegg=df_kegg.fillna('')
```

```python
#cleaning the keg columns
Drug_new=[]
for i in df_kegg['Drug']:
    a=i
    if re.search('\[[0-9A-z:]+\]',a):
        a=re.sub('\[[0-9A-z:]+\]','',a)
    Drug_new.append(a.strip())
Drug_new
```

```python
df_kegg['Drug_new']=Drug_new
```

```python
#cleaning the keg columns
Brite_new=[]
for i in df_kegg['Brite']:
    a=i
    if re.search('\[[0-9A-z:]+\]',a):
        a=re.sub('\[[0-9A-z:]+\]','',a)
    Brite_new.append(a.strip())
Brite_new
```

```python
df_kegg['Brite_new']=Brite_new
```

```python
#cleaning the keg columns
Gene_new=[]
for i in df_kegg['Gene']:
    a=i
    if re.search('\[[0-9A-z:]+\]',a):
        a=re.sub('\[[0-9A-z:]+\]','',a)
    Gene_new.append(a.strip())
Gene_new
```

```python
df_kegg['Gene_new']=Gene_new
```

```python
df_kegg.drop(['Brite','Gene','Brite'],axis=1,inplace=True)
```

```python
df_kegg
```

```python
name_caps_new=[]
for i in df_mayo['Name']:
    name_caps_new.append(i.upper())
df_mayo['Name_caps']=name_caps_new
df_mayo=df_mayo[['Name_caps', 'Overview_New', 'Symptoms_new', 'When_to_see_a_doctor_new',
       'Causes_new', 'Risk_factors_new', 'Complications_new',
       'Prevention_new']]
df_mayo.head()
```

```python
df_mayo.columns
```

```python
df_kegg_mayo_inner=pd.merge(df_kegg,df_mayo,how='inner',on=['Name_caps'])
df_kegg_mayo_inner
```

```python
df_kegg_mayo_outer=pd.merge(df_kegg,df_mayo,how='outer',on=['Name_caps'])
df_kegg_mayo_outer
```

```python
df_kegg_mayo=df_kegg_mayo_outer.append(df_kegg_mayo_inner)
```

```python
df_kegg_mayo.to_csv('kegg_mayo_consolidated.csv',index=False)
```
