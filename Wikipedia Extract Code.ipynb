Importing the libraries required for scraping the data

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

Loading webpage

```python
r = requests.get("https://en.m.wikipedia.org/wiki/List_of_infectious_diseases")
```

Fetching content

```python
s = bs(r.content)
```

```python
print(s.prettify())
```

```python
tab = s.find_all("table", class_="wikitable")[0]
tab
```

```python
cols = tab.find_all("th")
#cols = tab.find_all("strong")[0:5] (where th is not available)
```

```python
cols
```

Extracting column names 

```python
col_names = [x.text.strip('\n') for x in cols]
col_names
```

Looping through the table body, tbody, and table row, tr, to find the data within a table cell. 

```python
tab_rows = tab.find("tbody").find_all("tr")
tab_data=[]
for r in tab_rows:
    td = r.find_all('td')
    row = [str(r.get_text()).strip() for r in td] #stripping white spaces and html tags from the data
    tab_data.append(row)
tab_data
```

Converting data list into a dataframe

```python
df = pd.DataFrame(tab_data, columns = col_names)
df
```

Importing the dataframe to a csv file

```python
df.to_excel('Diseases_5.xlsx')
```
